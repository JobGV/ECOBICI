```python
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import seaborn as sns
from pySankey.sankey import sankey
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
warnings.simplefilter('ignore')
```

```python
ruta="https://raw.githubusercontent.com/JobGV/ECOBICI/main/Medidas_Centralidad/Centralidades_Consolidado.csv"
datos=pd.read_csv(ruta,encoding='latin1') 
```

# Outdregree

```python
cent = datos[(datos['mesure']=='out_deg_cent')]
n = 3
primeros_n=cent[(cent['Year']==2018) & (cent['Rank']<n)]
cent['Rank']=-cent.Rank

primeros_n=primeros_n['id'].tolist()
cent = cent[cent['id'].isin(primeros_n)]
ticks=cent.Rank.unique().tolist()
ticks_pos=cent.Rank.unique().tolist()
ticks_pos=cent.Rank.unique()*-1+1
ticks_pos = ticks_pos.tolist()

Ranking = 1+cent['Rank']*-1
Ranking = Ranking.tolist()
Anio = cent['Year'].tolist()
for i in range(len(Anio)):
    if Anio[i] == 2018:
        Anio[i] = 0
    elif Anio[i] == 2019:
        Anio[i] = 1
    elif Anio[i] == 2020:
        Anio[i] = 2
    elif Anio[i] == 2021:
        Anio[i] = 3
    else:
        Anio[i] = 4
        
        
EjeY = cent['Rank']
EjeY = EjeY.tolist()

aux = cent[cent['Year'] == 2018]
aux_id = aux['id']
aux_id =aux_id.tolist()
aux_id.sort()
Etiquetas = cent[cent['Year'] == 2022]
Estacion_Y = 1+(Etiquetas['Rank'])
Estacion_Y = Estacion_Y.tolist()
Estacion = Etiquetas['id']
Estacion =Estacion.tolist()
Estacion_nom = Etiquetas['nombre']
Estacion_nom =Estacion_nom.tolist()

colors = ['#009288','#d44787','#7343be','#2b2287']


custom_params = {"axes.spines.right": False, "axes.spines.top": False,"axes.spines.bottom": False,
                 "axes.spines.left": False}
sns.set_theme(style="ticks", rc=custom_params)
ax= sns.pointplot(x='Year', 
             y='Rank',
             hue='id',
             data=cent,
             palette=colors,
             scale=2.4)

plt.setp(ax.lines, alpha=.6)   

ax.set(xlabel='', ylabel='')
ax.xaxis.tick_top()
ax.tick_params(top=False)
ax.tick_params(left=False)
ax.tick_params(labelsize=18)

ax.set_yticks([])

for i in range(len(Anio)): 
    plt.text(Anio[i], EjeY[i]-0.1, Ranking[i],horizontalalignment='center',color='white',size='small',zorder=10)


Colors = []
for i in range(n):
    if Estacion[i] == aux_id[0]:
        Colors.append('#009288')
    elif Estacion[i] == aux_id[1]:
        Colors.append('#d44787')
    elif Estacion[i] == aux_id[2]:
        Colors.append('#7343be')
    else:
        Colors.append('#2b2287')
    

for i in range(n):
    plt.text(4.3, Estacion_Y[i]-1.1 ,Estacion_nom[i],horizontalalignment='left',color=Colors[i],size='medium',zorder=10)

sns.move_legend(ax, "upper left", bbox_to_anchor=(1, 1))
ax.legend([],[], frameon=False);

```
